## Part 1: Types of Machine Learning

### 1A: Supervised Learning Example
Supervised learning = model learns from **labelled** data (input + correct output).

Example: Predicting house price from size.

```python
# House size (sq ft) and prices
house_sizes = np.array([600, 800, 1000, 1200, 1500, 1800, 2000]).reshape(-1, 1)
house_prices = np.array([150000, 200000, 250000, 300000, 370000, 440000, 500000])

# Train a simple linear regression model
model = LinearRegression()
model.fit(house_sizes, house_prices)

# Predict price for a 1400 sq ft house
predicted = model.predict([[1400]])
print(f"Predicted price for 1400 sq ft: ${predicted[0]:,.0f}")

# Plot
plt.scatter(house_sizes, house_prices, color='blue', label='Actual')
plt.plot(house_sizes, model.predict(house_sizes), color='red', label='Prediction line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('Supervised Learning - Linear Regression')
plt.legend()
plt.show()
```

### Exercise 1A
1. Change the house sizes and prices to different values — does the prediction change?
2. Predict the price for a 2500 sq ft house.
3. Print `model.coef_` and `model.intercept_` — what do they mean?

```python
# 1. New data with different values
house_sizes_new = np.array([500, 700, 900, 1100, 1400, 1700, 2200]).reshape(-1, 1)
house_prices_new = np.array([120000, 180000, 230000, 280000, 350000, 420000, 520000])

model2 = LinearRegression()
model2.fit(house_sizes_new, house_prices_new)

# 2. Predict for 2500 sq ft
pred_2500 = model2.predict([[2500]])
print(f"Predicted price for 2500 sq ft: ${pred_2500[0]:,.0f}")

# 3. Coefficient and intercept
print(f"\nCoefficient (slope): {model2.coef_[0]:,.2f}")
print(f"Intercept: {model2.intercept_:,.2f}")
print("\nMeaning: For every 1 sq ft increase, price increases by the coefficient amount.")
print("Intercept is the base price when size = 0 (theoretical starting point).")
```

### 1B: Unsupervised Learning Example
Unsupervised learning = model finds **patterns on its own** — no labels given.

Example: Grouping customers by behaviour using KMeans Clustering.

```python
# Generate random customer data
X, _ = make_blobs(n_samples=150, centers=3, cluster_std=1.2, random_state=42)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title('Unsupervised Learning - KMeans Clustering (3 clusters)')
plt.legend()
plt.show()
```

### Exercise 1B
1. Change `n_clusters` to 2 and then 5 — what happens to the clusters?
2. Change `n_samples` to 300 — does it look different?
3. What is the difference between supervised and unsupervised learning in your own words?

```python
# 1. Try with 2 clusters
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, k in enumerate([2, 3, 5]):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    lbl = km.fit_predict(X)
    axes[i].scatter(X[:, 0], X[:, 1], c=lbl, cmap='viridis')
    axes[i].scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
                    s=200, c='red', marker='X')
    axes[i].set_title(f'K = {k} clusters')

plt.suptitle('Effect of changing n_clusters')
plt.tight_layout()
plt.show()

# 2. More samples
X_large, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.2, random_state=42)
km_large = KMeans(n_clusters=3, random_state=42, n_init=10)
lbl_large = km_large.fit_predict(X_large)

plt.scatter(X_large[:, 0], X_large[:, 1], c=lbl_large, cmap='plasma', alpha=0.6)
plt.title('300 samples - more data, clearer clusters')
plt.show()
```

**Difference between Supervised and Unsupervised:**

- **Supervised** → You tell the model the correct answers during training (labelled data). Model learns to map inputs → outputs. E.g., house size → price.
- **Unsupervised** → No labels given. The model discovers hidden patterns or groupings on its own. E.g., grouping customers by behaviour without telling it the groups.

### 1C: Reinforcement Learning — Concept
Reinforcement learning = an agent learns by **trial and error**, getting rewards for good actions.

Example: A robot learning to walk — every step forward = reward, falling = penalty.

## Part 2: Data Manipulation for Machine Learning

### 2A: Creating and Exploring a DataFrame

```python
# Student dataset
data = {
    'name': ['Neeraj', 'Arjun', 'Priya', 'Sara', 'Raj', 'Meera'],
    'age': [25, 28, 22, 30, np.nan, 26],
    'marks': [85, 90, 78, np.nan, 92, 88],
    'city': ['Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None],
    'passed': [True, True, True, False, True, True]
}

df = pd.DataFrame(data)
print(df)
print("\nShape:", df.shape)
print("\nNull values:\n", df.isnull().sum())
```

### 2B: Selecting, Filtering, and Sorting

```python
# Select a single column
print(df['marks'])

# Filter: students who passed
print("\nPassed students:")
print(df[df['passed'] == True])

# Sort by marks descending
print("\nSorted by marks:")
print(df.sort_values('marks', ascending=False))
```

### 2C: Handling Missing Values

```python
# Fill missing age with mean
df['age'] = df['age'].fillna(df['age'].mean())

# Fill missing marks with median
df['marks'] = df['marks'].fillna(df['marks'].median())

# Fill missing city with 'Unknown'
df['city'] = df['city'].fillna('Unknown')

print(df)
print("\nNull values after fill:\n", df.isnull().sum())
```

### 2D: Creating New Columns & GroupBy

```python
# Create a grade column
df['grade'] = df['marks'].apply(lambda x: 'A' if x >= 90 else ('B' if x >= 80 else 'C'))

# Average marks by city
print("Average marks by city:")
print(df.groupby('city')['marks'].mean())
print("\nFull dataframe:")
print(df)
```

```python
# 1. Add a new student
new_student = pd.DataFrame([{
    'name': 'Kiran', 'age': 24, 'marks': 95,
    'city': 'Bangalore', 'passed': True, 'grade': 'A'
}])
df = pd.concat([df, new_student], ignore_index=True)
print("After adding Kiran:")
print(df)

# 2. Filter Bangalore students
print("\nBangalore students:")
print(df[df['city'] == 'Bangalore'])

# 3. Student with highest marks
print("\nStudent with highest marks:")
print(df.loc[df['marks'].idxmax()])

# 4. Average age of grade A students
avg_age_A = df[df['grade'] == 'A']['age'].mean()
print(f"\nAverage age of Grade A students: {avg_age_A:.1f}")
```
